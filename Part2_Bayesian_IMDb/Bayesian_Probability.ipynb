# Part 2: Bayesian Probability

**1. Dataset Loading**

```python
import pandas as pd

file_path = "/content/IMDB Dataset.csv"

df = pd.read_csv(file_path, engine='python', on_bad_lines='skip')

reviews = df["review"].str.lower().tolist()
sentiments = df["sentiment"].str.lower().tolist()

total_reviews = len(reviews)
positive_reviews = [reviews[i] for i in range(total_reviews) if sentiments[i] == "positive"]

print("Total reviews:", total_reviews)
```

**2. Keyword Selection**

The following keywords were chosen by us basing on their common use in film reviews:

- **Positive keywords**: `great`and `wonderful` these are the traditionally used words with which the admiration and satisfaction are expressed and that is why they are strong signs of a positive review.

- **Negative keywords**: `terrible`and `boring`, these words are regularly used to express disappointment or dissatisfaction, thus, they serve as the powerful tools in expressing a bad review.

**3. Computing Prior**

> P(Positive)

```python
positive_count = sentiments.count("positive")
P_positive = positive_count / total_reviews

print("P(Positive):", P_positive)
```

**4. Likelihood Function**

> P(keyword | Positive)

```python
def likelihood(keyword):
    count = sum(1 for review in positive_reviews if keyword in review.split())
    return count / len(positive_reviews)
```

**5. Marginal Probability**

> P(keyword)

```python
def marginal(keyword):
    count = sum(1 for review in reviews if keyword in review.split())
    return count / total_reviews
```

**6. Posterior Using Bayes' Theorem**

> P(Positive | keyword) = P(keyword | Positive) × P(Positive) / P(keyword)

```python
def posterior(keyword):
    P_kw_pos = likelihood(keyword)
    P_kw = marginal(keyword)

    if P_kw == 0:
        return 0

    return (P_kw_pos * P_positive) / P_kw
```

**7. Computing for Keywords**

```python
keywords = ['great', 'wonderful', 'terrible', 'boring']

results = []
for word in keywords:
    results.append({
        "Keyword": word,
        "Prior P(Pos)": round(P_positive, 4),
        "Likelihood P(kw|Pos)": round(likelihood(word), 4),
        "Marginal P(kw)": round(marginal(word), 4),
        "Posterior P(Pos|kw)": round(posterior(word), 4),
    })

results_df = pd.DataFrame(results)
display(results_df)
```

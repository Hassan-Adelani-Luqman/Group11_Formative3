# Part 4: Gradient Descent in Code

This notebook implements the manual gradient descent calculations from Part 3.

## Problem Setup
- Linear equation: $y = mx + b$
- Initial parameters: $m_0 = -1$, $b_0 = 1$
- Learning rate: $\alpha = 0.1$
- Data points: $(1, 3)$ and $(3, 6)$
- Cost function: Mean Squared Error (MSE)

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array([1, 3])
y = np.array([3, 6])

# Initial parameters
m = -1.0
b = 1.0

# Learning rate
alpha = 0.1

# Number of data points
n = len(x)

print(f"Initial parameters: m = {m}, b = {b}")
print(f"Learning rate: α = {alpha}")
print(f"Data points: {list(zip(x, y))}")
```

## Iteration 1

Current parameters: $m = -1$, $b = 1$

```python
# Store history for plotting
m_history = [m]
b_history = [b]
error_history = []

print("=== Iteration 1 ===")
print(f"Current parameters: m = {m}, b = {b}")

# Predictions
y_pred_1 = m * x[0] + b
y_pred_2 = m * x[1] + b
print(f"\nPredictions:")
print(f"  ŷ₁ = {m}({x[0]}) + {b} = {y_pred_1}")
print(f"  ŷ₂ = {m}({x[1]}) + {b} = {y_pred_2}")

# Errors
e1 = y_pred_1 - y[0]
e2 = y_pred_2 - y[1]
print(f"\nErrors (ŷ - y):")
print(f"  e₁ = {y_pred_1} - {y[0]} = {e1}")
print(f"  e₂ = {y_pred_2} - {y[1]} = {e2}")

# MSE
mse = (e1**2 + e2**2) / n
error_history.append(mse)
print(f"\nMSE = ({e1}² + {e2}²) / {n} = {mse}")

# Gradients
grad_m = (e1 * x[0] + e2 * x[1]) / n
grad_b = (e1 + e2) / n
print(f"\nGradients:")
print(f"  ∂J/∂m = ({e1})({x[0]}) + ({e2})({x[1]}) / {n} = {grad_m}")
print(f"  ∂J/∂b = ({e1}) + ({e2}) / {n} = {grad_b}")

# Update parameters
m_old, b_old = m, b
m = m - alpha * grad_m
b = b - alpha * grad_b
m_history.append(m)
b_history.append(b)

print(f"\nUpdates:")
print(f"  m₁ = {m_old} - {alpha}({grad_m}) = {m}")
print(f"  b₁ = {b_old} - {alpha}({grad_b}) = {b}")
```

## Iteration 2

Current parameters from Iteration 1

```python
print("=== Iteration 2 ===")
print(f"Current parameters: m = {m}, b = {b}")

# Predictions
y_pred_1 = m * x[0] + b
y_pred_2 = m * x[1] + b
print(f"\nPredictions:")
print(f"  ŷ₁ = {m}({x[0]}) + {b} = {y_pred_1}")
print(f"  ŷ₂ = {m}({x[1]}) + {b} = {y_pred_2}")

# Errors
e1 = y_pred_1 - y[0]
e2 = y_pred_2 - y[1]
print(f"\nErrors (ŷ - y):")
print(f"  e₁ = {y_pred_1} - {y[0]} = {e1}")
print(f"  e₂ = {y_pred_2} - {y[1]} = {e2}")

# MSE
mse = (e1**2 + e2**2) / n
error_history.append(mse)
print(f"\nMSE = ({e1}² + {e2}²) / {n} = {mse}")

# Gradients
grad_m = (e1 * x[0] + e2 * x[1]) / n
grad_b = (e1 + e2) / n
print(f"\nGradients:")
print(f"  ∂J/∂m = ({e1})({x[0]}) + ({e2})({x[1]}) / {n} = {grad_m}")
print(f"  ∂J/∂b = ({e1}) + ({e2}) / {n} = {grad_b}")

# Update parameters
m_old, b_old = m, b
m = m - alpha * grad_m
b = b - alpha * grad_b
m_history.append(m)
b_history.append(b)

print(f"\nUpdates:")
print(f"  m₂ = {m_old} - {alpha}({grad_m}) = {m}")
print(f"  b₂ = {b_old} - {alpha}({grad_b}) = {b}")
```

## Iteration 3

```python
print("=== Iteration 3 ===")
print(f"Current parameters: m = {m}, b = {b}")

# Predictions
y_pred_1 = m * x[0] + b
y_pred_2 = m * x[1] + b
print(f"\nPredictions:")
print(f"  ŷ₁ = {m:.2f}({x[0]}) + {b:.2f} = {y_pred_1:.2f}")
print(f"  ŷ₂ = {m:.2f}({x[1]}) + {b:.2f} = {y_pred_2:.2f}")

# Errors
e1 = y_pred_1 - y[0]
e2 = y_pred_2 - y[1]
print(f"\nErrors (ŷ - y):")
print(f"  e₁ = {y_pred_1:.2f} - {y[0]} = {e1:.2f}")
print(f"  e₂ = {y_pred_2:.2f} - {y[1]} = {e2:.2f}")

# MSE
mse = (e1**2 + e2**2) / n
error_history.append(mse)
print(f"\nMSE = {mse:.4f}")

# Gradients
grad_m = (e1 * x[0] + e2 * x[1]) / n
grad_b = (e1 + e2) / n
print(f"\nGradients:")
print(f"  ∂J/∂m = {grad_m:.2f}")
print(f"  ∂J/∂b = {grad_b:.2f}")

# Update parameters
m_old, b_old = m, b
m = m - alpha * grad_m
b = b - alpha * grad_b
m_history.append(m)
b_history.append(b)

print(f"\nUpdates:")
print(f"  m₃ = {m_old:.4f} - {alpha}({grad_m:.2f}) = {m:.4f}")
print(f"  b₃ = {b_old:.4f} - {alpha}({grad_b:.2f}) = {b:.4f}")
```

## Iteration 4

```python
print("=== Iteration 4 ===")
print(f"Current parameters: m = {m:.4f}, b = {b:.4f}")

# Predictions
y_pred_1 = m * x[0] + b
y_pred_2 = m * x[1] + b
print(f"\nPredictions:")
print(f"  ŷ₁ = {m:.4f}({x[0]}) + {b:.4f} = {y_pred_1:.4f}")
print(f"  ŷ₂ = {m:.4f}({x[1]}) + {b:.4f} = {y_pred_2:.4f}")

# Errors
e1 = y_pred_1 - y[0]
e2 = y_pred_2 - y[1]
print(f"\nErrors (ŷ - y):")
print(f"  e₁ = {y_pred_1:.4f} - {y[0]} = {e1:.4f}")
print(f"  e₂ = {y_pred_2:.4f} - {y[1]} = {e2:.4f}")

# MSE
mse = (e1**2 + e2**2) / n
error_history.append(mse)
print(f"\nMSE = {mse:.6f}")

# Gradients
grad_m = (e1 * x[0] + e2 * x[1]) / n
grad_b = (e1 + e2) / n
print(f"\nGradients:")
print(f"  ∂J/∂m = {grad_m:.4f}")
print(f"  ∂J/∂b = {grad_b:.4f}")

# Update parameters
m_old, b_old = m, b
m = m - alpha * grad_m
b = b - alpha * grad_b
m_history.append(m)
b_history.append(b)

print(f"\nUpdates:")
print(f"  m₄ = {m_old:.6f} - {alpha}({grad_m:.4f}) = {m:.6f}")
print(f"  b₄ = {b_old:.6f} - {alpha}({grad_b:.4f}) = {b:.6f}")
```

## Final Results and Predictions

```python
print("\n" + "="*50)
print("FINAL RESULTS")
print("="*50)
print(f"\nFinal parameters after 4 iterations:")
print(f"  m = {m:.6f}")
print(f"  b = {b:.6f}")

print(f"\nFinal predictions using y = {m:.6f}x + {b:.6f}:")
for i in range(len(x)):
    y_pred = m * x[i] + b
    print(f"  x = {x[i]}: ŷ = {y_pred:.4f} (actual y = {y[i]}, error = {y_pred - y[i]:.4f})")

print(f"\nFinal MSE: {error_history[-1]:.6f}")

print("\n" + "="*50)
print("TREND ANALYSIS")
print("="*50)
print("\nParameter changes:")
print(f"  m: {m_history[0]} → {m_history[-1]:.6f} (converging towards optimal value)")
print(f"  b: {b_history[0]} → {b_history[-1]:.6f} (converging towards optimal value)")
print(f"\nError reduction: {error_history[0]} → {error_history[-1]:.6f}")
print("✓ Parameters are moving towards reducing the error with each iteration")
```

## Visualization: Parameter Evolution

Plot 1: Shows how parameters $m$ and $b$ change over iterations

```python
plt.figure(figsize=(12, 5))

# Plot 1: Parameters over iterations
plt.subplot(1, 2, 1)
iterations = range(len(m_history))
plt.plot(iterations, m_history, 'bo-', label='m (slope)', linewidth=2, markersize=8)
plt.plot(iterations, b_history, 'ro-', label='b (intercept)', linewidth=2, markersize=8)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Parameter Value', fontsize=12)
plt.title('Parameter Evolution over Iterations', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.xticks(iterations)

# Annotate final values
plt.annotate(f'm_final = {m_history[-1]:.3f}', 
             xy=(len(m_history)-1, m_history[-1]), 
             xytext=(10, -20), textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.5', fc='lightblue', alpha=0.7),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.annotate(f'b_final = {b_history[-1]:.3f}', 
             xy=(len(b_history)-1, b_history[-1]), 
             xytext=(10, 20), textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.5', fc='lightcoral', alpha=0.7),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.tight_layout()
plt.show()
```

## Visualization: Error Reduction

Plot 2: Shows how Mean Squared Error decreases over iterations

```python
plt.figure(figsize=(10, 5))

# Plot 2: Error over iterations
error_iterations = range(1, len(error_history) + 1)
plt.plot(error_iterations, error_history, 'go-', linewidth=2, markersize=10)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.title('Error Reduction over Iterations', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(error_iterations)

# Annotate error values
for i, error in enumerate(error_history):
    plt.annotate(f'{error:.4f}', 
                xy=(i+1, error), 
                xytext=(0, 10), textcoords='offset points',
                ha='center', fontsize=9,
                bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.6))

# Add initial vs final error text
plt.text(0.5, 0.95, f'Initial MSE: {error_history[0]:.4f}\nFinal MSE: {error_history[-1]:.6f}',
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7),
         fontsize=10)

plt.tight_layout()
plt.show()
```

## Visualization: Linear Regression Fit

```python
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(x, y, color='red', s=100, zorder=3, label='Data points')

# Plot initial line
x_line = np.linspace(0, 4, 100)
y_initial = -1 * x_line + 1
plt.plot(x_line, y_initial, 'r--', alpha=0.5, linewidth=2, label='Initial line (m=-1, b=1)')

# Plot final line
y_final = m * x_line + b
plt.plot(x_line, y_final, 'b-', linewidth=3, label=f'Final line (m={m:.3f}, b={b:.3f})')

# Plot true line (from the two points)
true_m = (y[1] - y[0]) / (x[1] - x[0])
true_b = y[0] - true_m * x[0]
y_true = true_m * x_line + true_b
plt.plot(x_line, y_true, 'g--', linewidth=2, alpha=0.7, label=f'Perfect fit (m={true_m:.1f}, b={true_b:.1f})')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Gradient Descent: Linear Regression Fit', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(-0.5, 4)
plt.ylim(-3, 7)

plt.tight_layout()
plt.show()
```
